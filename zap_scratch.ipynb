```python

from matplotlib import pyplot as plt
abf = zap_abf
sweep_num = 0
response_gain = 1
stim_gain = 10
offset = 0
response_chan_num = 1
stim_channel_num = 2
sweep_plot_list = [0, 10, 14]
fig = plt.figure(figsize=(8, 5))
clamp_mode = "CurrentClamp"

for i in sweep_plot_list:
    sweep_num = i
    abf.setSweep(sweep_num, channel=response_chan_num)
    # plot the ADC (voltage recording)
    ax1 = fig.add_subplot(211)
    ax1.set_title("ADC (recorded waveform)")
    response_vec = abf.sweepY*response_gain + offset
    ax1.plot(abf.sweepX, response_vec, alpha = .5)
    time_vec = abf.sweepX

    # plot the DAC (clamp current)
    ax2 = fig.add_subplot(212, sharex=ax1)  # <-- this argument is new
    ax2.set_title("DAC (stimulus waveform)")
    abf.setSweep(sweep_num, channel= stim_channel_num)
    #abf.setSweep(sweep_num, channel=0)
    stim_vec = abf.sweepY * stim_gain
    ax2.plot(abf.sweepX, stim_vec, alpha = .5)
    sweep = Sweep(t=time_vec,
              v=response_vec,
              i=stim_vec,
              sampling_rate=25000,
              sweep_number=0,
              clamp_mode=clamp_mode,
              #epochs = curr_epoch
              )

# decorate the plots
ax1.set_ylabel(abf.sweepLabelY)
ax2.set_xlabel(abf.sweepLabelX)
ax2.set_ylabel(abf.sweepLabelC)
#ax1.axes.set_xlim(0, .35)  # <-- adjust axis like this
plt.show()
```

```python
from ipfx import chirp
from ipfx.sweep import Sweep, SweepSet
from ipfx import epochs
```

```python
ss = list()
ss.append(sweep)
sweep_set = SweepSet(ss)
```

```python
(start_ind, end_ind) = epochs.get_stim_epoch(sweep.i)
print(start_ind)
sweep.i[0:10]
#sweep.i[end_ind+1]
1/(sweep.t[1] - sweep.t[0]) * .35
start_ind = 3500
end_ind = int(20 * 1/(sweep.t[1] - sweep.t[0]))
```

```python
(resistance, reactance, freq) = chirp.chirp_amp_phase(sweep_set, start = sweep.t[start_ind], end = sweep.t[end_ind], max_freq=20, min_freq = 1, down_rate = 100)
```

```python
fig = plt.figure(figsize=(8, 5))
plt.plot(freq, resistance, alpha = .5)
```

```python
from scipy.signal import savgol_filter
import numpy as np
```

```python
n_filt = int(np.rint(1/(freq[1]-freq[0])))*2 + 1
filt = lambda x: savgol_filter(x, n_filt, 3)
amp, phase = map(filt, [resistance, reactance])
```

```python
fig = plt.figure(figsize=(8, 5))
plt.plot(freq[2:-1], amp[2:-1], alpha = .5)

i_max = np.argmax(amp[2:-1])
z_max = freq[i_max]
print(z_max)
```
