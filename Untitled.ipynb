```python
import pyabf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import glob
```

```python
abf_file_name = '19320030.abf'

file_base_base_path = '/Users/stripathy/Downloads/valiante_lab_ephys_mar_2020/L5/Homeira/Total5Homeira-Last-Version - with 2020 Data/'

fn = file_base_base_path + abf_file_name
```

```python
# import gain and offset info

# read in cell metadata and processed ephys property dictionary
# this is needed because we need to know what the RMP values are per cell

l5_gain_info = '/Users/stripathy/rstudio_projects/valiante_ih/raw-data/final_kri_data/gain_offset_info/Layer 5_Homeira_Stimulus_Gain_Offset_ April 02_20.xlsx'
#excel_file = file_base_base_path + 'valiante_lab_ephys_mar_2020/L23/Homeira/Total2Homeira-Lastversion_March 10_2020.xlsx'
gain_info = pd.read_excel(open(l5_gain_info, 'rb'))  



```

```python
meta_row = gain_info.loc[gain_info['File name'] == abf_file_name]
offset = meta_row['Offset voltage'].values[0]
```

```python

abf = pyabf.ABF(fn)
#stim_channel_num = 3
#response_chan_num = 2

#stim_channel_num = 3
#response_chan_num = 2

stim_gain = .2
response_gain = .25

stim_end = 2

sweep_num = 0
sweep_plot_list = [0, 10, 15]
fig = plt.figure(figsize=(8, 5))

for i in sweep_plot_list:
    sweep_num = i
    abf.setSweep(sweep_num, channel=response_chan_num)
    # plot the ADC (voltage recording)
    ax1 = fig.add_subplot(211)
    ax1.set_title("ADC (recorded waveform)")
    ax1.plot(abf.sweepX, abf.sweepY*response_gain + offset, alpha = .5)

    # plot the DAC (clamp current)
    ax2 = fig.add_subplot(212, sharex=ax1)  # <-- this argument is new
    ax2.set_title("DAC (stimulus waveform)")
    abf.setSweep(sweep_num, channel=stim_channel_num)
    #abf.setSweep(sweep_num, channel=1)
    stim_vec = abf.sweepY * stim_gain
    ax2.plot(abf.sweepX, stim_vec, alpha = .5)

# decorate the plots
ax1.set_ylabel(abf.sweepLabelY)
ax2.set_xlabel(abf.sweepLabelX)
ax2.set_ylabel(abf.sweepLabelC)
ax1.axes.set_xlim(0, stim_end)  # <-- adjust axis like this
plt.show()
```
